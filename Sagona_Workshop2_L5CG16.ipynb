```python
a=pd.read_csv("/content/sample_data/medical_students_dataset (3).csv")
a
```

```python
a.info()
```

```python
print(a.isnull().sum())
```

```python
print(a.columns)
```

```python
# Numerical columns
a['Age'] = a['Age'].fillna(a['Age'].mean())
a['BMI'] = a['BMI'].fillna(a['BMI'].mean())
a['Cholesterol'] = a['Cholesterol'].fillna(a['Cholesterol'].mean())

# Categorical columns
a['Gender'] = a['Gender'].fillna(a['Gender'].mode()[0])
a['Blood Type'] = a['Blood Type'].fillna(a['Blood Type'].mode()[0])
a['Diabetes'] = a['Diabetes'].fillna(a['Diabetes'].mode()[0])
a['Smoking'] = a['Smoking'].fillna(a['Smoking'].mode()[0])
```

```python
print(a.isnull().sum())
```

```python
a.duplicated().sum()
```

```python
a = a.drop_duplicates()
```
